```python
with open('systematictheology.txt') as f:
    txt = [ t.rstrip() for t in f.readlines()]
```

```python
txt
```

```python
from fpdf import FPDF
```

```python
with open('sysTheo.txt', 'w') as f:
    for l in txt:
        if l == '':
            f.write('\n')
        elif '__________________________________________________________________' in l:
            l2 = l.strip('__________________________________________________________________')
            f.write(l2 + '\n')
        else:
            f.write(l.lstrip() + ' ')
```

```python
pdf = FPDF()
```

```python
pdf.output('sysTheo.txt', 'systematictheology.pdf')
```

```python
pdf.add_page()
```

```python
pdf.set_font("Arial", size = 15)
```

```python
with open('sysTheo.txt') as f2:
    txt2 = [l for l in f2.readlines()]
```

```python
pdf.cell(200, 10, txt = txt2, ln = 1, align = 'C')
```
